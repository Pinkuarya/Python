```python
# Q6) We will Compare the difference between all the years features with SalesPrice.

# Select numerical columns
numerical_columns = df.select_dtypes(include='number')

# Extract columns representing years and SalesPrice
years_columns = [col for col in numerical_columns.columns if 'Year' in col]
sales_price_column = 'SalePrice'

# Calculate the differences between each year feature and SalesPrice
for year_column in years_columns:
    difference = df[year_column] - df[sales_price_column]
    print(f"\nDifference between '{year_column}' and 'SalesPrice':")
    print(difference)
```

```python
# Q7) On the Discrete Variable Find the relationship between Discrete and Sales price.

discrete_columns = df.select_dtypes(include=['int64', 'object'])

# Plot relationship between each discrete variable and SalesPrice

for column in discrete_columns.columns:
    if column != 'SalePrice':  # Skip SalesPrice column itself
        plt.figure(figsize=(4, 2))
        sns.boxplot(x=column, y='SalePrice', data=df)
        plt.title(f'Relationship between {column} and SalePrice')
        plt.xlabel(column)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
# Q8) On the Continous Variable Find the relationship between Discrete and Sales Price.

continuous_columns = df.select_dtypes(include= 'float64')

# Plot relationship between each continuous variable and SalesPrice
for column in continuous_columns.columns:
    plt.figure(figsize=(4, 2))
    sns.scatterplot(x=column, y='SalePrice', data=df)
    plt.title(f'Relationship between {column} and SalePrice')
    plt.xlabel(column)
    plt.ylabel('SalePrice')
    plt.show()
```

```python
# Q9) Analyse the Continous values by creating the histogram to understand the distribution.

continuous_columns = df.select_dtypes(include='float64')

# Plot histograms for each continuous variable

for column in continuous_columns.columns:
    plt.figure(figsize=(5, 3))
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Q10) Apply the Lograthmic Transformation.

continuous_columns = df.select_dtypes(include='float64')

# Apply logarithmic transformation to each continuous variable
for column in continuous_columns.columns:
    df[column] = np.log1p(df[column])  # Applying log transformation using np.log1p() to avoid log(0) errors

# Plot histograms for each transformed continuous variable
for column in continuous_columns.columns:
    plt.figure(figsize=(5, 3))
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of Log-transformed {column}')
    plt.xlabel(f'Log-transformed {column}')
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Q1) Find the Outliers.

# Define quartiles (Q1, Q3), interquartile range (IQR), and threshold
Q1 = numerical_columns.quantile(0.25)  # 25th percentile
Q3 = numerical_columns.quantile(0.75)  # 75th percentile

IQR = Q3 - Q1  # Interquartile range

threshold = 1.5  # Threshold multiplier

# Find outliers
outliers = (numerical_columns < (Q1 - threshold * IQR)) | (numerical_columns > (Q3 + threshold * IQR))

# Print columns with outliers
columns_with_outliers = outliers.any()
print("Columns with outliers:")
print(columns_with_outliers[columns_with_outliers])
```

```python
# Q5) Do the Feature Engineering for the
# 		a) Handle the Missing Values
# 		b) Handle the Categorical Variables
# 		c) Handle  the Numerical Variables
# 		d) Handling  the Temporal Varaibles

# a) Handle the Missing Values
# First check for missing values, and there are no missing values in this code.
numerical_columns = df.select_dtypes(include='number')
print(numerical_columns.isnull().sum())

# If there were missing values, impute the missing values along with the mean for the numerical variable.
# df[numerical_columns.columns] = df[numerical_columns.columns].fillna(df[numerical_columns.columns].mean())

# For categorical variables, impute missing values with mode
categorical_columns = df.select_dtypes(include='object')

if not categorical_columns.empty:
    print("Categorical columns with missing values:", categorical_columns.columns[categorical_columns.isnull().any()].tolist())
    df[categorical_columns.columns] = df[categorical_columns.columns].fillna(df[categorical_columns.columns].mode().iloc[0])

# b) Handle the Categorical Variables
# Convert categorical variables using one-hot encoding
if not categorical_columns.empty:
    df = pd.get_dummies(df, columns=categorical_columns.columns)

# Update categorical columns after one-hot encoding
categorical_columns = df.select_dtypes(include='uint8')
print("Updated categorical columns:", categorical_columns.columns)
print("Number of updated categorical columns:", len(categorical_columns.columns))

# c) Handle the Numerical Variables
# Scale numerical variables using Min-Max scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[numerical_columns.columns] = scaler.fit_transform(df[numerical_columns.columns])

# d) Handle the Temporal Variables (if applicable)
# Extract year, month, day from temporal variable
# Replace 'Timestamp' with the correct name of the temporal variable in your dataset
if 'Timestamp' in df.columns:
    df['Year'] = df['Timestamp'].dt.year
    df['Month'] = df['Timestamp'].dt.month
    df['Day'] = df['Timestamp'].dt.day
    df = df.drop('Timestamp', axis=1)  # Drop the original temporal variable

# Ensure all missing values are handled
df = df.fillna(0)  # Fill any remaining missing values with 0
```
